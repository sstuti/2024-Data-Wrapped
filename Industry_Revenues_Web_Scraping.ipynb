```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## 2024 Company Revenues

```python
URL = "https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue"
page = requests.get(URL).text

soup = BeautifulSoup(page, "html.parser")
```

```python
results = soup.find(id = "mw-content-text")
```

```python
table = results.find("table")
```

```python
df_2024 = pd.DataFrame(columns=['Company_Name', 'Industry', 'Revenue ($ million)', 'Headquarters'])

for row in table.tbody.find_all('tr'):    
    columns = row.find_all('td')
    
    if(columns != []):
        company_name = columns[0].text.strip()
        industry = columns[1].text.strip()
        revenue = columns[2].text.strip()
        hq = columns[5].text.strip()
        
        new_row = pd.DataFrame([{
                'Company_Name': company_name,
                'Industry': industry,
                'Revenue ($ million)': revenue,
                'Headquarters': hq
                }])

        df_2024 = pd.concat([df_2024, new_row], ignore_index=True)
```

```python
df_2024['Revenue ($ million)'] = df_2024['Revenue ($ million)'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype('float')
```

```python
df_2024.loc[df_2024['Company_Name'] == 'Enel', 'Revenue ($ million)'] = 147100
```

## 2004 Company Revenues

```python
URL = "https://fortune.com/ranking/global500/2004/"
page = requests.get(URL).text

soup = BeautifulSoup(page, "html.parser")
```

```python
results = soup.find(id = "search-section")
```

```python
table = results.find("table")
```

```python
# Industry and Headquarters filled in after research
df_2004 = pd.DataFrame(columns=['Company_Name', 'Industry', 'Revenue ($ million)', 'Headquarters'])

for row in table.tbody.find_all('tr'):    
    columns = row.find_all('td')
    
    if(columns != []):
        company_name = columns[1].text.strip()
        revenue = columns[2].text.strip()
        
        new_row = pd.DataFrame([{
                'Company_Name': company_name,
                'Revenue ($ million)': revenue,
                }])

        df_2004 = pd.concat([df_2004, new_row], ignore_index=True)
```

```python
df_2004 = df_2004.loc[:49]
```

```python
df_2004['Revenue ($ million)'] = df_2004['Revenue ($ million)'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype('float')
```
